```python
def get_camera_counts(text):
    if pd.isna(text):
        return pd.Series([np.nan, np.nan])
    
    text = text.lower()

    # Split by '&' into rear and front parts
    parts = [p.strip() for p in text.split('&')]
    
    # Assign parts
    rear_part = parts[0] if len(parts) > 0 else ''
    front_part = parts[1] if len(parts) > 1 else ''

    # Rear camera regex checks
    if re.search(r'\bquad\b', rear_part):
        rear = 4
    elif re.search(r'\btriple\b', rear_part):
        rear = 3
    elif re.search(r'\bdual\b', rear_part):
        rear = 2
    elif 'rear' in rear_part:
        rear = 1
    else:
        rear = 0

    # Front camera regex checks
    if re.search(r'\bquad\b', front_part):
        front = 4
    elif re.search(r'\btriple\b', front_part):
        front = 3
    elif re.search(r'\bdual\b', front_part):
        front = 2
    elif 'front' in front_part:
        front = 1
    else:
        front = 0

    return pd.Series([rear, front])

def extract_primary_mp(text):
    if pd.isna(text):
        return pd.Series([np.nan, np.nan])

    text = text.lower()
    parts = [p.strip() for p in text.split('&')]
    
    rear_part = parts[0] if len(parts) > 0 else ''
    front_part = parts[1] if len(parts) > 1 else ''

    # Extract first MP value from rear and front parts
    rear_match = re.search(r'(\d+)\s*mp', rear_part)
    front_match = re.search(r'(\d+)\s*mp', front_part)

    rear_mp = int(rear_match.group(1)) if rear_match else 0
    front_mp = int(front_match.group(1)) if front_match else 0

    return pd.Series([rear_mp, front_mp])
```

```python
# Again extracting all the camera details as this time we have the camera details of all the phones
df[["num_rear_cameras", "num_front_cameras"]] = df['camera_specs'].apply(get_camera_counts)
```

```python
df[["rear_primary_mp", "front_primary_mp"]] = df['camera_specs'].apply(extract_primary_mp)
```

```python
def extract_screen_type(specs):
    specs = re.sub(r"\s+", " ", specs)
    parts = [part.strip() for part in specs.split(",")]
    if len(parts) > 1:
        match = re.search(
            r"([A-Za-z0-9\s\+\-\.]+)\s+screen",
            specs,
            re.IGNORECASE
        )
        if match:
            return match.group(1).upper().strip()
    
    return np.nan
```

```python
df['screen_type'] = df['display_specs'].apply(extract_screen_type)
```

```python
def clean_screen_type(screen):
    if pd.isna(screen):
        return np.nan

    screen = screen.upper().strip()

    # SUPER AMOLED and premium AMOLED types
    if "SUPER AMOLED" in screen or "DYNAMIC AMOLED" in screen or "FLUID AMOLED" in screen or "SUPER FLUID AMOLED" in screen:
        return "SUPER AMOLED"
    
    # Standard AMOLED
    if "AMOLED" in screen:
        return "AMOLED"
    
    # OLED that’s not AMOLED
    if "OLED" in screen:
        return "OLED"
    
    # All LCD-like screens
    if "LCD" in screen or "TFT" in screen or "LTPS" in screen or "IPS" in screen:
        return "LCD"
    
    return "OTHER"
```

```python
df["screen_type"] = df['screen_type'].apply(clean_screen_type)
```

```python
grouped_mode = df.groupby(['brand', 'price_group'])['screen_type'].agg(lambda x: x.mode().iloc[0] if not x.mode().empty else np.nan)
price_group_mode = df.groupby('price_group')['screen_type'].agg(lambda x: x.mode().iloc[0] if not x.mode().empty else np.nan)

def impute_missing_screen_type(row):
    val = grouped_mode.get((row['brand'], row['price_group']), np.nan)
    if pd.isna(val):
        return price_group_mode.get(row['price_group'], np.nan)
    return val

missing_screen_type_df = df[df['screen_type'].isnull()]
missing_screen_type_df['screen_type'] = missing_screen_type_df.apply(impute_missing_screen_type, axis=1)
```

```python
df.update(missing_screen_type_df)
```
